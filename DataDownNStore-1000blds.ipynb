```python
dvr = pd.read_parquet(county1000, storage_options={"anon": True},columns=['timestamp',
                                                                          'bldg_id',col_name])
dvr.columns = ['timestamp','out_total_energy']

new  = pd.DataFrame(index=dvr['timestamp'][0:4*24*365])

blds = list(set(dvr.index))
                    
for i in range(len(blds)):
    bld= blds[i]
    new[bld] = np.asarray(dvr[dvr.index==bld]['out_total_energy'],dtype='float16')
new.to_csv('county1000.csv')
```
